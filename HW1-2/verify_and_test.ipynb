### Create test json 

```python
import os 
import json

popular_list = list(os.walk('./popular_1'))
popular_test = [popular_list[0][0] + '/' + x for x in popular_list[0][2]]

not_popular_list = list(os.walk('./not_popular_1'))
not_popular_test = [not_popular_list[0][0] + '/' + x for x in not_popular_list[0][2]]
```

```python
# random select 10000 images from popular and not popular
import random
popular_test = random.sample(popular_test, 10000)
not_popular_test = random.sample(not_popular_test, 10000)
```

```python
# dump to json
json_data = {
    "image_paths": popular_test + not_popular_test
}
with open('test.json', 'w') as f:
    json.dump(json_data, f)
```

### Verify the result

```python
import json

result = json.load(open('image_predictions.json', 'r'))['image_predictions']
answer = [1] * 10000 + [0] * 10000

list_len = len(result)

path = json.load(open('test.json', 'r'))['image_paths']

# calculate accuracy
acc_list = [1 if result[i] == answer[i] else 0 for i in range(list_len)]
accuracy = sum(acc_list) / list_len
print(f'accuracy: {accuracy}, {sum(acc_list)} are correct, {list_len - sum(acc_list)} are wrong')
print(f'Wrong pictures')
for i in range(list_len):
    if result[i] != answer[i]:
        print(f'{path[i]}')
```
